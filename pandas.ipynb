## Analytical Summary of the dataset

```python
covid.describe(include='all')
```

## Drop irrelevant column

```python
covid = covid.drop(covid.iloc[:, 5:14], axis=1)
covid.head(5)
```

```python
covid = covid.drop(['Country/Region'], axis=1)
covid
```

Rename Column

```python
covid = covid.rename(columns={"WHO Region":"Regions"})
```

```python
covid.shape
```

```python
covid.hist(figsize=(20,30))
```

## Relationship between Categorical and continuous variable

```python
sns.boxplot(x='Regions', y='Recovered', data=covid.head(20))
```

```python
sns.pairplot(covid)
```

```python
# Checks Rows has Cointains Duplicate data or not
duplicate_rows_covid = covid[covid.duplicated()]
duplicate_rows_covid.shape
```

```python
# if Duplicate is there then drop this by...drop_duplicate method
covid = covid.drop_duplicates()
covid.head()
```

```python
# find the null values
print(covid.isnull().sum())
```

```python
# Drop the missing Values
covid = covid.dropna()
```

```python
covid.count()
```

```python
covid.Regions.value_counts().plot(kind='bar', figsize=(10, 5))
plt.title('Number of Active cases in the Regions ')
plt.ylabel('Number of active cases')
plt.xlabel('Who Regions');
```

```python
# Find the relations between the variables.
cor=covid.corr()
sns.heatmap(cor, cmap='BrBG', annot=True)
cor
```
